# Jadval yaratish, ma'lumot qo'shish

```python
query = """

CREATE TABLE IF NOT EXISTS UZB(
  ind INT,
  viloyat TEXT PRIMARY KEY,
  zichlik DOUBLE,
  aholi DOUBLE NOT NULL
);

"""
cursor.execute(query)
db.commit()
```

```python
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
cursor.fetchall()
```

```python
pd.read_sql("SELECT * FROM UZB", db)
```

```python
cursor.execute("PRAGMA table_info(UZB);")
cursor.fetchall()
```

```python
query = "INSERT INTO UZB VALUES (1, 'Xorazm', 333.333, 5555.5) "
cursor.execute(query)
db.commit()
```

```python
cursor.execute("SELECT * FROM UZB")
cursor.fetchall()
```
