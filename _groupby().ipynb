# Guruhlash. .groupby()

.groupby() metodi yordamida jadvaldagi ma'lumotlarni biror ustun qiymaltlari guruhlarga ajratib tahlil qilishimiz mumkin

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data.csv", index_col=0)
df.head()
```

```python
df.company.values
```

Yuqoridagi jadvalda turli avtomobil brandlari haqida ma'lumotlar jamlangan. Keling boshlanishiga har bir brand uchun ustunlarning o'rtacha qiymatini topamiz:

```python
df.groupby('company').mean()
```

Yuqoridagi natijani tahlil qilsak, `alfa-romero` mashinalarning o'rtacha narxi (`price`) `15498$`, `honda` mashinalarining o'rtacha ot kuchi (`horsepower`) `92.3` ga teng ekan.

Bitta ustun uchun qiymatlarni olish ham mumkin:

```python
df.groupby('company').price.min()
```

```python
df.groupby(['company','num-of-cylinders']).mean()
```

.groupby() yordamida nafaqat o'rta balki boshqa qiymatlarni ham hisblash mumkin

```python
df.groupby('company').max()
```

.agg() motodi yordamida birdaniga bir nechta statik ma'lumotlarni olishimiz ham mumkin:

```python
df.groupby('company').agg([len, max, min, lambda x: x*2]).head()
```

```python
df.groupby('company').agg([np.mean, max, len]).head()
```
