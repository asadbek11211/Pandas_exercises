Keling df1 ga yana cator kiritamiz Suvning ifloslanishi va Havoning tozaligi haqidagi collumns larni qo'shamiz

```python
df1['AirPolution'] = 100 - df1.AirQuality
df1['WaterQuality'] = 100 - df1.WaterPolution
```
