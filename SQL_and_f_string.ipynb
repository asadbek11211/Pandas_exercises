# *SQL ga f string yordamida murojat qilish*

```python
import pandas as pd
import sqlite3 as sql
```

```python
db = sql.connect("mydb.db")
cursor = db.cursor()
```

```python
df = pd.read_csv("sample_data/california_housing_train.csv")
df.head()
```

```python
query = """
  CREATE TABLE California_hose(
    house_age INT,
    house_room INT,
    house_badrooms INT,
    house_pop INT
  );
"""
cursor.execute(query)
db.commit()
```

```python
query = "SELECT name FROM sqlite_master WHERE type='table';"
cursor.execute(query)
cursor.fetchall()
```

```python
pd.read_sql(query, db)
```

```python
df1 = df[df.total_bedrooms == 2]
df1[['housing_median_age','total_rooms','total_bedrooms','population']]
```

df ning ixtiyoriy indeksdagi ma'lumotlarni db ga **saqlaymiz**

```python
ind = df1.index.values[0]
age = df.at[ind,'housing_median_age']
rooms = df.at[ind,'total_rooms']
bedrooms = df.at[ind,'total_bedrooms']
population = df.at[ind,'population']

query = f"""
  INSERT INTO California_hose VALUES ({age}, {rooms}, {bedrooms}, {population})
"""
cursor.execute(query)
db.commit()
```
