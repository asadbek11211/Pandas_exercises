```python
import numpy as np
import pandas as pd

#    
data = pd.read_csv('/content/Data.txt', delim_whitespace=True, header=None)

#     NumPy
z = np.array(data[0])
s = np.array(data.iloc[:, 1:])
```

          .      Matplotlib.

```python
import matplotlib.pyplot as plt

#  
plt.figure(figsize=(8, 6))
for i in range(s.shape[1]):
    plt.plot(s[:, i], z, label=f'sensor {i+1}')
plt.xlabel('Signal')
plt.ylabel('Depth')
plt.legend()
plt.show()
```
