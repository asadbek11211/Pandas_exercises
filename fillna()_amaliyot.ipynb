# Ma'lumotlarni to'ldirish

Agar imkoni bo'lsa mavjud bo'lmagan qiymatlarni to'ldirish tavsiya qilinadi. Buning uchun .fillna() metodiga murojat qilamiz

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data_nan.csv", index_col=0)
```

```python
df
```

Quyidagi misolda barcha NaN qiymatlarni 0 bilan almashtiramiz (garchi bunday qilish tavsiya qilinmasada)

```python
df1 = df.fillna(0)
df1
```

quyidagi misolda wheel-base va horsepower ustunlari uchun alohida standart qiymatlar beramiz

```python
df2 = df.fillna({'wheel-base':67, 'horsepower':90})
df2
```

method='ffill' parametri yordamida NaN qiymatlarni avvalgi qataordagi qiymatlar bilan to'ldirishimiz mumkin. Agar jadvalga ahamiyat bersak, ba'zi qavtolar qayta-qayta takrorlangan (narxi turlicha xolos). Demak bir avto haqidagi ma'lumotni o'xshash avto ma'lumotidan olishimiz mantiqan to'g'ri.



```python
df3 = df[df.company == 'audi']
df3
```

```python
df3.fillna(method='ffill')
```

method='bfill' yordamida esa ekasincha, NaN qiymatni o'zidan keyingi qiymat bilan to'ldirib ketasiz.



```python
df4 = df[df.company == 'volkswagen']
df4
```

```python
df4.fillna(method='bfill')
```

limit parametri yordamida aynan nechta NaN qiymatlar ketma-ketligini to'ldirishni ham ko'rsatish mumkin. Tasavvur qiling, biror qatorda 10-20 ta qator qiymatlar yo'q shunday holatda bularning hammasini emas, boshida 2-3tasini avvalgi qator bilan to'ldirish talab qilinda df.fillna(method='ffill', limit=2) deb ko'rsatasiz.


Bo'sh qiymatlarni to'ldirishning yana bir usuli ustun uchun o'rta qiymatni olish. Quyidagi misolda hatchback turidagi moshinalar uchun mavjud bo'lmagan qiymatlarni o'rta qiymat bilan to'ldiramiz.



```python
df5 = df[df.company == 'mazda']
df5
```

```python
df5.fillna(method = 'ffill', limit=5)
```

```python
df6 = df[df.company == 'toyota']
df6
```

```python
df6.fillna(df6.mean())
```

```python
df.fillna({'length':np.mean(df.length)})
```
