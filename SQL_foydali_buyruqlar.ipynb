# BETWEEN, IN VA LIKE

```python
query = "SELECT * FROM elect  WHERE country='Uzbekistan' AND  year BETWEEN 1995 AND 2000"
pd.read_sql(query, db)
```

```python
query = "SELECT * FROM elect WHERE country='Uzbekistan' and year BETWEEN 1990 AND 2000"
pd.read_sql(query, db)
```

```python
query = "SELECT * FROM elect WHERE country='Uzbekistan' and year IN (1990, 1992, 2000, 2003, 2009) ;"
pd.read_sql(query, db)
```

# Like(qidirish) kamandasi
% -> %var = elsevar, anvar, sarvar

_ -> _ar = car, bar, nar

__ -> __rom = ikrom, 

%ol% -> nolam, bolam 

```python
query = "SELECT * FROM elect WHERE country LIKE '%stan' AND year IN (1990, 1995, 2000)"
pd.read_sql(query, db)
```

```python
query = "SELECT * FROM elect WHERE country LIKE '%bek%' AND year IN (1990, 1995, 2000)"
pd.read_sql(query, db)
```

```python
query = "SELECT * FROM elect WHERE country LIKE '%na%' AND year IN (1990, 1995, 2000)"
pd.read_sql(query, db)
```

```python
query = "SELECT * FROM elect WHERE code LIKE 'A__' AND year IN (1990, 1995, 2000)"
pd.read_sql(query, db)
```

```python
query = "SELECT * FROM elect WHERE country LIKE 'R%' AND year IN (1990, 1995, 2000)"
pd.read_sql(query, db)
```
