# .replace() va .rename() - qiymatlarini almashtirish

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data_nan.csv", index_col=0)
df.head()
```

.replace() funksiyasi df ning qiymatlariga qayta o'zgartirish kiritadi

```python
df.replace('four',4)
```

```python
df.replace({'two': 2, 'six': 6})
```

.rename() df ning index va collumns larining qiymatlarini qayta kiritishda qo'laniladi

```python
df.rename(columns=str.upper)
```

```python
indeks = df.company.values
indeks
```

```python
df.index = indeks
```

```python
df
```

```python
df.rename(index=str.upper, columns=str.upper)
```

```python
df = df.rename(index=str.capitalize, columns=str.capitalize)
df
```

```python
df.rename(columns={'Average-mileage': 'milage'}, inplace=True)
df
```
