# Mavjud bo'lmagan qiymatlar bilan ishlash *dropna()*

```python
import pandas as pd
```

Yangi df yaratamiz

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data_nan.csv", index_col=0)
df.head()
```

Keling, DF da mavjud bo'lmagan (NAN) qiymatlar sonini topamiz

.isnull()

.isnull() metodi yordamida ustun qiymati NaN ga teng (TRUE) yoki yo'q (FALSE) ekanini tekshiramiz. Ya'ni NaN qiymat o'rnida TRUE ko'ramiz

```python
df.isnull()
```

NaN qiymatlar sonini ko'ramiz .sum() metodi yordamida

```python
df.isnull().sum()
```

Keling bu ko'rsatkichlarni jami ma'lumotlarining foizi sifatida chiqaramiz.

```python
son = df.isnull().sum() 
son1 = len(df)
(son/son1) * 100
```

pandas da NaN qiymatlar bilan ishlash un bir nechta qiymatlar bor:

  dropna - NaN qiymatlar bor ustun yoki qatorlarni o'chirib taxshlaydi

  fillna - NaN qiymatlarni boshqa qiymatlar bilan to'ldirib ketish(misol uchun avvalgi yoki keyingii qiymatlar uchun)

  isnull - NaN qiymatlarni filtirlash (NaN qiymat TRUE qaytaradi)

  notnull - isnull ning aksi

```python
df1 = df.dropna()
df1
```

```python
df1.shape
```

```python
df2 = df.dropna(axis=1)
df2
```

```python
df2.shape
```

thresh parametri yordamida biz NaN bo'lmagan qiymatlarining soniga chegara belgilashimiz mumkin. Misol uchun, quyidagi kodimiz NaN bo'lmagan qiymatlar 48 dan kam bo'lgan ustunlarni tashlab yuboradi.



```python
df.dropna(axis=1, thresh=48)
```

subset parametri yordamida aynan qaysi ustunlarni tekshirishni ham ko'rsatish mumkin. Quyidagi misolda faqatgina length va body-style ustunlarida NaN qiymati bor qatorlarni tashlab yuboramiz.



```python
df3 = df.dropna(subset=['length', 'body-style'])
df3
```

```python
df3.shape
```

```python
df4 = df.dropna(subset=['horsepower'])
df4
```
