```python
db = sql.connect("chinook.db")
cursor = db.cursor()
```

```python
query = "SELECT * FROM customers"
df = pd.read_sql(query, db)
df.head()
```

```python
query = "SELECT lastName, FirstName, Address, Country FROM customers"
df = pd.read_sql(query, db)
df.head()
```

```python
query = 'SELECT * FROM customers WHERE country="USA"'
pd.read_sql(query, db)
```

```python
query = 'SELECT * FROM customers WHERE country="UZB" OR country="Brazil"'
pd.read_sql(query, db)
```

```python
query = 'SELECT * FROM customers WHERE country="UZB" OR country="Brazil" ORDER BY lastName '
pd.read_sql(query, db)
```
