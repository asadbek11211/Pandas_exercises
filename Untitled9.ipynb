```python
import matplotlib.pyplot as plt
import numpy as np
from fake_zond import generate_fake_zond
from latest_version_5_part_belousov_by_potapovs_fixes import ks

epsilon = 0.01


def eta(params, forward, indx):
    params_1 = params.copy()
    params_1[indx] = params_1[indx] * (1 + epsilon)
    print(params)
    print(params_1)
    return (1 / epsilon) * (forward(*params_1) / forward(*params) - 1)


def draw(params, titles):
    fig, axs = plt.subplots(nrows=1, ncols=3)
    graphs = [axs[0], axs[1], axs[2]]

    for i in range(len(graphs)):
        # if eta:
        #     draw_for_one(*params, graphs[i], titles[i], eta, i)
        draw_for_one(*params[i], graphs[i], titles[i])
    plt.show()


# args = z, rho1, rho2, z0, lengths_to_zonds
def draw_for_one(*args):
    for i in args:
        if type(i) == type(np.arange(1, 10, 1)):
            param = i

    for i in range(len(args[4])):
        zond = generate_fake_zond(args[4][i], name=f'Zond with length {args[4][i]}')
        # if args[7]:
        #     args[5].plot(eta([args[0], args[1], args[2], args[3]], zond, args[8] + 1), args[0], label=zond.__doc__)

        args[5].plot(zond(args[0], args[1], args[2], args[3]), param, label=zond.__doc__)
    args[5].set_title(args[6])
    args[5].invert_yaxis()
    args[5].legend()

def draw_eta(params, names):
    fig, axs = plt.subplots(nrows=1, ncols=3)
    graphs = [axs[0], axs[1], axs[2]]
    lengths = [0.3, 1.0]
    for i in range(len(graphs)):
        for l in lengths:
            zond = generate_fake_zond(l, name=f'Zond FAKE with length {l}')
            graphs[i].plot(eta(params[:-1], zond, i+1), params[0], label=zond.__doc__)
        graphs[i].set_title(names[i])
        graphs[i].legend()
        graphs[i].invert_yaxis()
    plt.show()

def draw_eq(names, zond):
    # for g in range(len(names)):
    rho1 = np.arange(-200, 200, 1)
    data = np.zeros((100, 100))
    # zond = generate_fake_zond(2, name='Zond FAKE with length 2')
    l = np.linspace(0.1, 10, 100)
    rho1 = np.linspace(10, 150, 100)
    rho2 = np.linspace(10, 250, 100)
    z = np.linspace(0, 100, 100)
    for i in range(len(rho1)):
        for j in range(len(rho2)):
            zond = generate_fake_zond(l[j], name='Zond FAKE with length 2')
            data[i, j] = zond(z[i], 10, 100, 51)
    print(data)
    # fig, axs = plt.subplots(1, 2, figsize=(10, 3), constrained_layout=True)
    plt.pcolor(l, z, data)
    plt.colorbar()
    plt.show()

if __name__ == '__main__':
    '''Тута пилим зонды'''
    # params np.arange(0, 101, 1), 10, 200, 50, [1, 0.3, 5]
    params_to_z = [np.arange(-2, 2, 0.01), 10, 100, 0.01, [5.0, 2.0, 0.3]]
    params_to_rho1 = [51, np.arange(0, 101, 1), 100, 50, [5.0, 2.0, 0.3]]
    params_to_rho2 = [51, 10, np.arange(0, 101, 1), 50, [5.0, 2.0, 0.3]]
    # params_to_z0 = [51, 10, 200, np.arange(0, 101, 1), [5, 2, 0.3]]

    params = [params_to_z, params_to_rho1, params_to_rho2]  # нету params_to_z0
    titles = ['for z', 'for rho1', 'for rho2']  # нету 'for z0'

    # params = [params_to_z]
    # titles = ['for z']
    draw(params, titles)

    '''Тута зонды сравниваем'''
    params_to_fake = [np.arange(0, 101, 0.1), 100, 200, 50, 2]
    params_to_ks = [np.arange(50, -51, -0.1), 100, 200, 0, 2]
    fig, ax = plt.subplots()
    zond_fake = generate_fake_zond(params_to_fake[4], name='Zond FAKE with length 2')
    zond_ks = ks(params_to_ks[4], name='Zond KS with length 2')
    ax.plot(zond_fake(*params_to_fake[:-1]), params_to_fake[0], label=zond_fake.__doc__)
    ax.plot(zond_ks(*params_to_ks[:-1]), np.arange(101, 0, -0.1), label=zond_ks.__doc__)
    ax.legend()
    ax.invert_yaxis()
    plt.show()

    '''Тута исследуем чувствительность'''
    names = ['to rho1', 'to rho2', 'to z0']
    params = [np.arange(-2, 2, 0.01), 10, 100, 0.1, 2]
    draw_eta(params, names)  # оценка чувствительности относительно z0 нуждается в дополнительном анализе

    '''Тута исследуем эквивалентность'''
    names = ['rho1/rho2', 'rho1/l', 'rho2/l']
    draw_eq(names, zond_fake)
```
